```python
#add new item into inventory 

pid= input("enter product id : ")
if(pid not in rec):
    nm= input("enter product name : ")
    pr= int( input("enter product price : "))
    qn= int(input("enter product quantity : "))
    dis= float( input("enter % dicount on product  : "))
    expry= input("enter product expiry : ")
    tx= float(input("enter taxes product: "))

    rec[pid] = {"name":nm, "price":pr, "quantity" : qn, "disc" : dis, "exp" :expry, "tax" :tx }
else:
    print("duplicate entry")
```

```python
# code to add quantity

pid= input("enter product id : ")
if pid in rec:
    qn= int(input("enter product quantity : "))
    quan=rec[pid]["quantity"]+qn
    rec[pid]["quantity"]=quan
else:
    print("product not found")
```
