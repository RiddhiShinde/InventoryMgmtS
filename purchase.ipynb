```python
import json

fd=open("rec.json",'r')
r=fd.read()
fd.close()

rec= json.loads(r)
```

```python
rec
```

```python
m='y'
bill=0
while(m=='y'):
    
    print("~~~~~~~~~welcome~~~~~~~~")
    print("available inventory: ")
    for nm in rec.items():
        print(nm)
    print(end='\n')
    print("~~~~~~~~~~~~~~~~~~~~~~~~~~")
    print(end='\n')
    
    pid=input("enter product id: ")   
    ui_qua= int(input("enter quantity: ")) 
     
    if (pid not in rec):
        print("product id not  found !!!")
    else:
        
        if (rec[pid]["quantity"]<1):
            print("product not available")
            break
            
        else:
            
            if (rec[pid]["quantity"]<ui_qua) :
                print("sorry. we have only left", rec[pid]["quantity"])
                p=input("do you want to purchase that much? (y/n)")
                if (p=='n'):
                    break       

                print("***************")
                print("product detail :")
                print("product name : ", rec[pid]["name"])
                print("product price : ",rec[pid]["price"])
                print("---------------")
                bill+=rec[pid]["price"] * ui_qua
                print("total bill:", bill)

                #update record
                rec[pid]["quantity"] -= ui_qua

                print("***************")
                m=input("do you want to purchase more product ? (y/n)")
                print(end='\n')

                if(m=='n'):
                    print("thank you ..")
                    break
```
